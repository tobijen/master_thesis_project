## Build qrels object

```python
qrels = (
    qrels
    .rename(columns={
        "base_query_id": "query_id",
        "document_id": "doc_id",
        "judgment_score": "relevance"
    })
)

qrels
```

## Build run objects and evaluate

```python
# Aufteilen der Resultsdataframes in Subdataframes. Filtern nach (search-method, query_type, document_typ)
# Jedes Subdataframe besteht aus query_id, document_id, score

from ir_measures import calc_aggregate, nDCG, P, R, RR, AP

measures = [
    nDCG@10,
    P(rel=2)@5,
    RR(rel=2),
]

rows = []

for (method, qvar, dvar), sub in search_results.groupby(
    ["search_method", "query_type", "document_type"]
):
    run_df = sub[
        ["base_query_id", "document_id", "score"]
    ]

    run_df = run_df.rename(columns={"base_query_id": "query_id", "document_id": "doc_id"})

    # nur relevante Qrels für diese Queries
    qrels_sub = qrels[qrels["query_id"].isin(run_df["query_id"])]

    res = calc_aggregate(measures, qrels_sub, run_df)

    rows.append({
        "search_method": method,
        "query_type": qvar,
        "document_type": dvar,
        "nDCG@10": res[nDCG@10],
        "P@5": res[P(rel=2)@5], # ab label 2 relevant
        "MRR": res[RR(rel=2)], # ab label 2 relevant
    })

evaluation_results = pd.DataFrame(rows)
```

```python
evaluation_results[["nDCG@10", "P@5", "MRR"]] = (
    evaluation_results[["nDCG@10", "P@5", "MRR"]]
    .round(4)
)
```

```python
evaluation_results
```

### Add duration

```python
avg_duration = (
    search_results
        .groupby(["search_method", "query_type", "document_type"])["duration"]
        .mean()
        .reset_index()
        .rename(columns={"duration": "avg_duration"})
)

avg_duration
```

```python
evaluation_results_with_duration = evaluation_results.merge(
    avg_duration,
    on=["search_method", "query_type", "document_type"],
    how="left"
)
```

```python
evaluation_results_with_duration
```

```python
# save evaluation results
evaluation_results_with_duration.to_csv(
    EVALUATION_DATA + "evaluation_results.csv",
    mode="w",
    header=True,
    index=False,
    sep=";",
    decimal=","
  )
```
